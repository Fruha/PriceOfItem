```python
import requests as req
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gdown
import ast
import json
import urllib

from tqdm.notebook import tqdm
```

```python

gdown.download('https://drive.google.com/uc?export=download&id=1mHqIAH7wzC1yi1ezPfxJVkRLLku6cGeH', 'ABOBA.csv', quiet=False)
data = pd.read_csv('ABOBA.csv')
data
```

```python
# %%time
def temp_func(x):
    overprice_float = 0
    if pd.isna(x['overpay']) == False:
        overprice_float = ast.literal_eval(x['overpay']).get('float', 0)
    return overprice_float
temp = data.apply(temp_func, axis=1)
temp.name = 'overprice_float'
temp
```

```python
data['overprice_float'] = temp
```

```python
def temp_func(x):
    price = x['price']
    overprice = 0
    if x['userOverprice'] == x['userOverprice']:
        price -= x['userOverprice']
    try:
        overprice = ast.literal_eval(x['overpay']).get('stickers', 0)
        if overprice < price:
            price -= overprice
    except:
        pass
    try:
        overprice = ast.literal_eval(x['overpay']).get('float', 0)
        if overprice < price:
            price -= overprice
    except:
        pass
    return price

data.loc[:,['realPrice']] = data.apply(temp_func, axis=1)
data.sort_values('realPrice', ascending=False).head(30)
```

```python
data = data[data['overprice_float'] != 0]
```

```python
data.loc[:,'OV_F/RP'] = data['overprice_float'] / data['realPrice']
```

```python
def temp_func(x):
    return 'https://steamcommunity.com/market/listings/730/' + urllib.parse.quote(x['fullName'])
data.loc[:,'steamUrl'] = data.apply(temp_func, axis=1)
data
```

```python
data[data['realPrice'] > 10][pd.isna(data['isStatTrak']) == True].sort_values(['OV_F/RP'], ascending=False).drop_duplicates(subset=['fullName'])
```
