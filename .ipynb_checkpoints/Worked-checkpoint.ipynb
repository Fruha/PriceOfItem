#Work with data

##Add real price

```python
def temp_func(x):
    price = x['price']
    overprice = 0
    if x['userOverprice'] == x['userOverprice']:
        price -= x['userOverprice']
    try:
        overprice = ast.literal_eval(x['overpay']).get('stickers', 0)
        if overprice < price:
            price -= overprice
    except:
        pass
    try:
        overprice = ast.literal_eval(x['overpay']).get('float', 0)
        if overprice < price:
            price -= overprice
    except:
        pass
    return price

data.loc[:,['realPrice']] = data.apply(temp_func, axis=1)
data.sort_values('realPrice', ascending=False).head(10)
```

```python
data.head()
```

##Data selecting

```python
t = data.loc[:,['realPrice','nameId','overprice','price','quality','stickers','overpay','fullName','isStatTrak','isSouvenir','isStatTrak']]
t.head()
```

##Add overpay_stickers

```python
def temp_func(x):
    try:
        return ast.literal_eval(x['overpay'])['stickers']
    except:
        return 0

t['overpay_stickers'] = t.apply(temp_func, axis = 1)
t.head(5)
```

##Remove skins without stickers

```python
 def temp_func(x):
    if (x['stickers'] == x['stickers']):
        return ast.literal_eval(x['stickers'].replace("\'","\""))
    else:
        return np.nan

t1 = t[t.apply(lambda x: x['stickers'] == x['stickers'], axis = 1)].reset_index()
t1.head()
```

##Change quality

```python
def temp_func(x):
    temp_dict = {
        'bs' : 'Battle-Scarred',
        'fn' : 'Factory New',
        'ft' : 'Field-Tested',
        'mw' : 'Minimal Wear',
        'ww' : 'Well-Worn'
    }
    return temp_dict[x['quality']]

t1.loc[:,['quality']] = t1.apply(temp_func,axis=1)
t1.head()
```

##Add weaponName

```python
def temp_func(x):
    ans = x['fullName']
    ans = ans.split('|')[0]
    ans = ans.replace('StatTrak™', '')
    ans = ans.replace('Souvenir', '')
    ans = ans.strip()
    # (Factory New)
    # (Minimal Wear)
    # (Field-Tested)
    # (Well-Worn)
    # (Battle-Scarred)

    return ans
t1['weaponName'] = t1.apply(temp_func, axis=1)
t1.head()
```

## Deleting wear stickers and unpack stickers + generate addintional feature

```python
def temp_func(x):
    stickers = ast.literal_eval(x['stickers'])
    columns = ['name','overprice','price','wear','count']
    counts = dict()
    ans = np.full([len(columns)*5], np.nan).astype('object')
    for sticker in stickers:
        if sticker != None:
            position = sticker.get('position', np.nan)
            for i1, column in enumerate(columns):
                ans[position*len(columns) + i1] = sticker.get(column, np.nan)
            if ans[position*len(columns) + columns.index('wear')] != 0:
                for i1, column in enumerate(columns):
                    ans[position*len(columns) + i1] = np.nan
    for name in ['overprice', 'price']:
        for i in range(5):
            if pd.isna(ans[i*len(columns) + columns.index(name)]):
                ans[i*len(columns) + columns.index(name)] = 0.0
    for name in ans[columns.index('name')::len(columns)]:
        counts[name] = counts.get(name, 0) + 1
    for i in range(5):
        name = ans[i*len(columns) + columns.index('name')]
        if not pd.isna(name):
            ans[i*len(columns) + columns.index('count')] = counts.get(ans[i*len(columns) + columns.index('name')], 0)
        else:
            ans[i*len(columns) + columns.index('count')] = 0
    return ans
t2 = t1.iloc[:,:].apply(temp_func, axis = 1)
t2 = pd.DataFrame(t2.tolist(), columns=[f'{name}_{i}' for i in range(5) for name in ['name','overprice','price','wear','count']])
t2.head()
```

## Create database_prices

```python
database_prices_weapon = t1.groupby(f'fullName')['realPrice'].min()
database_prices_weapon
```

```python
database_prices_sticker = pd.Series()
for i in range(5):
    temp = t2.groupby(f'name_{i}')[f'price_{i}'].min()
    database_prices_sticker = pd.concat([database_prices_sticker, temp], axis=0)
database_prices_sticker = database_prices_sticker.groupby(database_prices_sticker.index).first()
database_prices_sticker = database_prices_sticker.rename(lambda x: x.replace('Sticker | ',''))
database_prices_sticker
```

##Merging to bots_skins

```python
bots_skins = pd.concat([t1,t2], axis=1)
bots_skins = bots_skins[bots_skins['isSouvenir'] != bots_skins['isSouvenir']]
bots_skins.head(5)
```

##Spliting data

```python
%%time
def splitData(X):
    ans = pd.DataFrame()
    for i in tqdm(range(5)):
        temp = X.copy()
        temp['sticker№'] = i
        temp['overprice_sticker'] = temp[f'overprice_{i}']
        ans = pd.concat([ans,temp], ignore_index=True)
    ans = ans.drop(columns=['index'])
    ans['sticker№'] = ans['sticker№'].astype('str')
    return ans
```

```python
url = 'steam://rungame/730/76561202255233023/+csgo_econ_action_preview%20M4425333723162582825A22847867902D9371076196844472113'

def get_api_url(url):
    if url.find('S') != -1:
        m = url[url.find('S')+1:url.find('A')]
        a = url[url.find('A')+1:url.find('D')]
        d = url[url.find('D')+1:]
        return f'https://api.csgofloat.com/?m={m}&a={a}&d={d}'
    else:
        s = url[url.find('M')+1:url.find('A')]
        a = url[url.find('A')+1:url.find('D')]
        d = url[url.find('D')+1:]
        return f'https://api.csgofloat.com/?s={s}&a={a}&d={d}'

get_api_url(url)
```

```python
def get_price_of_stickers(url):
    data = req.get(url).json()['iteminfo']
    # return data
    # clear
    X_ = pd.DataFrame([{'realPrice' : 0.0,
                        'quality' : 'nan',
                        'fullName' : 'nan',
                        'isSouvenir' : 'nan',
                        'weaponName' : 'nan',
                        'name_0' : 'nan',
                        'price_0' :	0.0,
                        'wear_0' : float('nan'),
                        'count_0' : 0,
                        'name_1' : 'nan',
                        'price_1' :	0.0,
                        'wear_1' : float('nan'),
                        'count_1' : 0,
                        'name_2' : 'nan',
                        'price_2' :	0.0,
                        'wear_2' : float('nan'),
                        'count_2' : 0,
                        'name_3' : 'nan',
                        'price_3' : 0.0,
                        'wear_3' : float('nan'),
                        'count_3' : 0,
                        'name_4' : 'nan',
                        'price_4' :	0.0,
                        'wear_4' : float('nan'),
                        'count_4' : 0,
                        }])
    try:
        X_['realPrice'] = database_prices_weapon[data['full_item_name']]
    except:
        X_['realPrice'] = -1
    for name in ['Factory New',  'Minimal Wear', 'Field-Tested', 'Well-Worn', 'Battle-Scarred']:
        if data['full_item_name'].find(name) != -1:
            X_['quality'] = name
            break
    X_['fullName'] = data['full_item_name']
    if X_['fullName'][0].find('Souvenir') != -1:
        return (X_['realPrice'],0)
    X_['isSouvenir'] = 'nan'
    X_['weaponName'] = X_['fullName']
    X_['weaponName'] = X_['weaponName'][0].split('|')[0].replace('StatTrak™', '').replace('Souvenir', '').strip()
    for sticker in data['stickers']:
        slot = sticker['slot']
        X_[f'name_{slot}'] = sticker['name']
        # print(database_prices_sticker[database_prices_sticker.index == sticker['name']].values[0])
        # print(sticker['name'])
        # print(database_prices_sticker[database_prices_sticker.index.str.find(sticker['name']) != -1].values)
        X_[f'price_{slot}'] = database_prices_sticker[database_prices_sticker.index == sticker['name']].values[0]
        if sticker.get('wear', 0) != 0:
            X_[f'wear_{slot}'] = np.nan
            continue
        else:
            X_[f'wear_{slot}'] = float(sticker.get('wear',0))
    for i in range(5):
        if X_[f'wear_{i}'][0] != X_[f'wear_{i}'][0]:
            X_[f'name_{i}'] = 'nan'
            X_[f'price_{i}'] = 0.0
    counts = dict()
    for i in range(5):
        name = X_[f'name_{i}'][0]
        counts[name] = counts.get(name, 0) + 1
    for i in range(5):
        name = X_[f'name_{i}'][0]
        if not pd.isna(name):
            X_[f'count_{i}'] = counts.get(name, 0)
        else:
            X_[f'count_{i}'] = 0
    ans = check_price_weapon(X_, -1)

    return (X_, X_['realPrice'][0], ans)


url = 'steam://rungame/730/76561202255233023/+csgo_econ_action_preview%20S76561198337602677A22886576017D5045872896882875692'

ans = get_price_of_stickers(get_api_url(url))
print('Weapon price:', ans[1])
print('Stickers price:', ans[2][0])
for i in range(5):
    print(f'Sticker slot {i}:', ans[2][i+1])
ans[0]
```

#Parse inventory

```python
id = 76561198337602677

def parse_inv(id):
    req_inv = req.get(f'https://steamcommunity.com/inventory/{id}/730/2?l=russian&count=1000')
    req_inv.json()
    try:
        items = req_inv.json()['descriptions']
    except:
        time.sleep(10)
        return pd.DataFrame()
    assets = pd.DataFrame(req_inv.json()['assets'])
    i = 0
    weapons_ = []
    for item in items:
        try:
            success = False
            for descr in item['descriptions']:
                num = descr['value'].find('Наклейка')
                if num != -1:
                    success = True
            if success == False:
                continue
            link = item['actions'][0]['link']
            link = link.replace('%owner_steamid%', f'{id}')
            class_id = item['classid']
            asset_id = assets[assets['classid'] == f'{class_id}']['assetid'].values[0]
            link = link.replace('%assetid%', f'{asset_id}')
            weapons_.append(link)
            i+=1
        except:
            pass
    print(i)
    data_inv = []

    for link in tqdm(weapons_, leave=False):
        try:
            data_inv.append(get_price_of_stickers(get_api_url(link)))
        except:
            pass
    data_inv
    temp = dict()
    temp['Weapon name'] = []
    temp['Weapon quality'] = []
    temp['Weapon price'] = []
    temp['Stickers price'] = []
    for i in range(5):
        temp[f'Name sticker slot {i}'] = []
        temp[f'Sticker slot {i}'] = []
    for i in data_inv:
        try:
            temp['Stickers price'].append(i[2][0])
            for i1 in range(5):
                temp[f'Name sticker slot {i1}'].append(i[0][f'name_{i1}'].values[0])
                temp[f'Sticker slot {i1}'].append(i[2][i1+1])
            temp['Weapon price'].append(i[1])
            temp['Weapon name'].append(i[0]['fullName'].values[0])
            temp['Weapon quality'].append(i[0]['quality'].values[0])
            temp['Stickers price/Weapon price']
        except:
            pass

    t = pd.DataFrame(temp).sort_values('Stickers price', ascending=False)
    t = t[t['Weapon quality'] != 'nan']
    return t

parse_inv(id)
```
